```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
d1 = np.random.binomial(1,.5, 100)
plt.axvline(d1.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(d1.mean() + d1.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(d1.mean() - d1.std(), color='b', linestyle='dashed', linewidth=2)

plt.hist(d1)
plt.show()
```

```python
#d1 shows that a binomial distribution will result in a distribution with two values. The Mean shows us that 0 is the
#more likely value for this random distribution. The standard deviation is less informative here.



d2 = np.random.chisquare(1,100)

plt.axvline(d2.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(d2.mean() + d2.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(d2.mean() - d2.std(), color='b', linestyle='dashed', linewidth=2)

plt.hist(d2)
plt.show()
```

```python
#d2 - the mean has value here showing the average value.

d3 = np.random.logistic(0,1,100)

plt.axvline(d3.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(d3.mean() + d3.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(d3.mean() - d3.std(), color='b', linestyle='dashed', linewidth=2)

plt.hist(d3)
plt.show()
```

```python
#In d3 the standard deviation shows us that the data is normally distributed.

d4 = np.random.lognormal(0,.1,100)

plt.axvline(d4.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(d4.mean() + d4.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(d4.mean() - d4.std(), color='b', linestyle='dashed', linewidth=2)

plt.hist(d4)
plt.show()
```

```python
#In d4 the standard deviation shows us that the data is normally distributed.

d5 = np.random.gamma(5,1, 100)

plt.axvline(d5.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(d5.mean() + d5.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(d5.mean() - d5.std(), color='b', linestyle='dashed', linewidth=2)

plt.hist(d5)
plt.show()
```

```python
#In d5 the standard deviation shows us that the data is not normally distributed.

d6 = np.random.poisson(3, 1000)

plt.axvline(d6.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(d6.mean() + d6.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(d6.mean() - d6.std(), color='b', linestyle='dashed', linewidth=2)

plt.hist(d6)
plt.show()
```

```python
#In d6 the standard deviation shows us that the data is normally distributed.
```

```python
#Generate two normally-distributed variables, one with a mean of 5 and standard deviation of 0.5, and the other with 
#a mean of 10 and standard deviation of 1.

mean = 5 
sd = .5
n = 1000

df = np.random.normal(mean, sd, 1000)

mean2 = 10 
sd2 = 1
n2 = 1000

df2 = np.random.normal(mean2, sd2, 1000)

#Add them together to create a third variable.

mean3 = mean + mean2
sd3 = sd + sd2
n = 1000

df3 = np.random.normal(mean3, sd3, 1000)

#Graph the third variable using a histogram.
#Compute the mean and standard deviation and plot them as vertical lines on the histogram.

plt.axvline(df3.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(df3.mean() + d6.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(df3.mean() - df3.std(), color='b', linestyle='dashed', linewidth=2)

plt.hist(df3)
plt.show()
```
