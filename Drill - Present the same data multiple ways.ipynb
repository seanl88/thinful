```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
%matplotlib inline

hit = pd.read_csv('Batting stats.csv')
hit.head(5)
```

```python
# Choose one variable and plot that variable four different ways.
# Plotting Batting Average by year

plotsize = (18.7, 8.27)
fig, ax = plt.subplots(figsize=plotsize)
sns.set_style("whitegrid")
sns.barplot(data = hit, x = 'Year', y = 'BA')
plt.title('Batting Average by Year')
plt.xticks(rotation=80)

fig, ax = plt.subplots(figsize=plotsize)
sns.set_style("whitegrid")
sns.pointplot(data = hit, x = 'Year', y = 'BA')
plt.title('Batting Average by Year')
plt.xticks(rotation=80)

plt.show()
```

```python
plotsize = (18.7, 8.27)
fig, ax = plt.subplots(figsize=plotsize)
ba= hit['BA']
sns.distplot(ba, hist=False, color="g", kde_kws={"shade": True})

plt.show()
```

```python
sns.lmplot(x="Year", y="BA", data=hit)
plt.title('Batting Average by Year')
plt.xticks(rotation=0)

plt.show()
```

```python
# Choose two continuous variables, and plot them three different ways.

crypto = pd.read_csv('crypto-markets.csv')
crypto.head(5)
```

```python
bitcoin = crypto[crypto['symbol'] == 'BTC']
bitcoin.head(5)
```

```python
roundedv = bitcoin.round({'volatility': 1})
roundedv.head(5)
```

```python
g = sns.lmplot(y='volatility', # Variable 1.
               x='close', # Variable 2.
               data=bitcoin, # Data
               fit_reg=False, # If set to true, plots a regression line.
               scatter_kws={'alpha':0.4}) # Set points to semi-transparent to see overlaping points.
g.set_ylabels("Volatility")
g.set_xlabels("Closing Price")
sns.plt.title('Volatility and Closing Price for Bitcoin')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=plotsize)
h = sns.boxplot(y='close', 
               x='volatility',
               data=roundedv)
sns.plt.title('Volatility and Closing Price for Bitcoin + Regression')


fig, ax = plt.subplots(figsize=plotsize)
sns.set_style("whitegrid")
sns.violinplot(data=roundedv, x='volatility', y ='close')
plt.title('Volatility and Closing Price for Bitcoin')
plt.xticks(rotation=90)

plt.show()
```

```python
valuelist = ['BTC', 'ETH',]
btc_eth = crypto[crypto.symbol.isin(valuelist)]
btc_eth.head(5)
```

```python
a = sns.lmplot(y='open',
               x='close',
               hue='symbol',
               data=btc_eth,
               fit_reg=False,
               scatter_kws={'alpha':0.4})
a.set_ylabels("Opening Price")
a.set_xlabels("Closing Price")
sns.plt.title('Volatility and Closing Price for Bitcoin and Ethereum')

b = sns.lmplot(y='open', 
               x='close',
               hue='symbol',
               data=btc_eth, 
               fit_reg=True,
               scatter_kws={'alpha':0.4})
b.set_ylabels("Opening Price")
b.set_xlabels("Closing Price")
sns.plt.title('Volatility and Closing Price for Bitcoin and Ethereum + Regression')

sns.set(style="whitegrid")
c = sns.factorplot(x="date", y="open", hue="symbol", data=btc_eth,
                   size=9, kind="point", palette="pastel",ci=95,dodge=True,join=False)
c.despine(left=True)
c.set_ylabels("Open")
c.set_xlabels("Date")
sns.plt.title('Pointplot: Open and Close for BTC and ETH')

plt.show()
```

```python
p = sns.FacetGrid(btc_eth, col='symbol')
p.map(plt.scatter,'open','close')
plt.show()
```

```python
vol_var = btc_eth[['variance','volatility','symbol']].groupby('symbol').mean()
vol_var.reset_index()
sns.boxplot(x="volatility", y="variance", data=vol_var, palette="PRGn")
plt.show()
```

```python
sns.set(style="whitegrid")

g = sns.factorplot(x="variance", y="volatility", hue="symbol", data=btc_eth,
                   size=6, kind="point", palette="pastel",ci=95,dodge=True,join=False)
g.despine(left=True)
g.set_ylabels("Volatility")
g.set_xlabels("Variance")
sns.plt.title('Pointplot: Variance and Volatility of BTC and ETH')
plt.show()
```
